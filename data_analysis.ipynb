```python
data_path = 'Data/spam.csv'
data = pd.read_csv(data_path, encoding='Windows-1252')
print(data[:10])
```

```python
data = data.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis=1)
data.columns = ['label', 'email']
```

```python
spam = data[data['label'] == 'spam']
ham = data[data['label'] == 'ham']
```

```python
def getEmailLength(df):
    total = 0
    for email in df['email']:
        total += len(email)
    return total / len(df)
```

```python
def frequentWords(df):
    words = []
    for email in df['email']:
        for word in str.split(email):
            words.append(word.lower())
    count = Counter(words)
    return count.most_common
```

```python
print(f'Average spam length: {getEmailLength(spam)}')
print(f'Average ham length: {getEmailLength(ham)}')
```

```python
print(frequentWords(spam[:10]))
print(frequentWords(ham[:10]))
```
